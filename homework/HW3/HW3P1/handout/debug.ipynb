```python
def InitializePaths(y_probs0, symbol_set):
        """
        Initialize paths with each of the symbols
        
        Input
        -----

        y_probs0 [np.array, dim=(len(symbols) + 1, batch_size)]
			batch size for part 1 will remain 1, but if you plan to use your
			implementation for part 2 you need to incorporate batch_size

        Returns
        -------
        
        InitialPathsWithFinalBlank, 
        InitialPathsWithFinalSymbol, 
        InitialBlankPathScore,
        InitialPathScore

        """
        path = '-'

        InitialBlankPathScore = {}
        InitialPathsWithFinalBlank = set()

        InitialBlankPathScore[path] = y_probs0[0]
        InitialPathsWithFinalBlank.add(path)

        InitialPathScore = {}
        InitialPathsWithFinalSymbol = set()

        for i, c in enumerate(symbol_set):
            path = c
            # beacuse of black in index 0
            InitialPathScore[path] = y_probs0[i + 1]
            InitialPathsWithFinalSymbol.add(path)
        
        return InitialPathsWithFinalBlank, InitialPathsWithFinalSymbol, \
                InitialBlankPathScore, InitialPathScore
```

```python
symbol_set = ["a", "b", "c"]
y_probs0 = [0.4155, 0.1450, 0.0860, 0.3535]
```

```python
InitialPathsWithFinalBlank, InitialPathsWithFinalSymbol, InitialBlankPathScore, InitialPathScore = InitializePaths(y_probs0, symbol_set)
```

```python
print("InitialPathsWithFinalBlank: ", InitialPathsWithFinalBlank)
print("InitialPathsWithFinalSymbol: ", InitialPathsWithFinalSymbol)
print("InitialBlankPathScore: ", InitialBlankPathScore)
print("InitialPathScore", InitialPathScore)
```
